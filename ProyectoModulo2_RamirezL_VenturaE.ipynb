### Portafolio 2017 2020

```python
import pandas_datareader.data as web
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

def getclose(names, start , end):
    return yf.download(names, start=start, end=end)["Adj Close"]

data_aapl = getclose("AAPL", "2017-01-01", "2020-12-31")
data_bofa = getclose("BAC", "2017-01-01", "2020-12-31")
data_coca = getclose("KO", "2017-01-01", "2020-12-31")
data_amex = getclose("AXP", "2017-01-01", "2020-12-31")
data_kraft = getclose("KHC", "2017-01-01", "2020-12-31")
data_tesla = getclose("TSLA", "2017-01-01", "2020-12-31")

Portafolio = data_aapl * 0.476 + data_bofa * 0.136 + data_coca * 0.072 + data_amex * 0.075 + data_kraft * 0.035 + data_tesla * 0.206

plt.figure(figsize=(12, 6))
plt.plot(Portafolio.index, Portafolio, label='Valor del Portafolio')
plt.xlabel('Fecha')
plt.ylabel('Valor del Portafolio')
plt.title('Valor del Portafolio de 2017 a 2020')
plt.legend()
plt.show()
```

### Rendimientos

```python
rendimientos_portafolio = Portafolio.pct_change()
maximo = rendimientos_portafolio.max()
minimo = rendimientos_portafolio.min()
media = rendimientos_portafolio.mean()
desviacion_estandar = rendimientos_portafolio.std()
print("Máximo:", maximo)
print("Mínimo:", minimo)
print("Media:", media)
print("Desviación Estándar:", desviacion_estandar)

```


### Simulacion para 2021


```python
num_simulaciones = 1000 
num_dias = 252 
simulaciones = pd.DataFrame()

for i in range(num_simulaciones):
    rendimientos_aleatorios = np.random.normal(media_rendimientos, desviacion_estandar_rendimientos, num_dias)
    valor_portafolio_simulado = (1 + rendimientos_aleatorios).cumprod() * Portafolio[-1]
    simulaciones[f'Simulación {i+1}'] = valor_portafolio_simulado

plt.figure(figsize=(12, 6))
plt.plot(simulaciones)
plt.xlabel('Días')
plt.ylabel('Valor del Portafolio')
plt.title('Simulación de Montecarlo para el año 2021')
plt.show()
```

### Probabilidad de que en 2021 suba un 20% o más.

```python
simulaciones = np.random.normal(media_rendimientos, desviacion_estandar_rendimientos, size=(num_simulaciones, 1)) + 1
portafolio_2021 = Portafolio[-1] * np.cumprod(simulaciones)

probabilidad = np.sum(portafolio_2021 > Portafolio[-1] * 1.2) / num_simulaciones

print("La probabilidad de que el portafolio suba más del 20% en 2021 es: %",probabilidad * 100)
```
